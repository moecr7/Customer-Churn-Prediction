```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("churndata.csv")

print(df['Churn'].value_counts(normalize=True))

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count plot of the target column 'Churn'
sns.countplot(x='Churn', data=df)

# Set plot title
plt.title('Target Distribution')

# Show the plot
plt.show()
```

```python
# Select only numeric columns
numerics = df.select_dtypes(include='number').columns

# Plot histograms for all numeric features
df[numerics].hist(figsize=(12, 10), bins=30)

# Add a title to the whole figure
plt.suptitle('Distributions of Numerical Features')

# Show the plots
plt.show()
```

```python
# Create a boxplot of MonthlyCharges grouped by Churn status
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)

# Show the plot
plt.title('Monthly Charges by Churn')
plt.show()
```

```python
# Count plot of Contract types, segmented by Churn status (Yes/No)
sns.countplot(x='Contract', hue='Churn', data=df)

# Add title for better readability
plt.title('Churn Count by Contract Type')
plt.show()
```
